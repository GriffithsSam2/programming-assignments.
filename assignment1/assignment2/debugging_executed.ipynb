```python
def id_to_fruit(fruit_id: int, fruits: set[str]) -> str:
    # 1. Convert the unordered set into a sorted list so the order is predictable
    fruits_list = sorted(list(fruits))
    
    # 2. Grab the fruit at the requested index directly
    if 0 <= fruit_id < len(fruits_list):
        return fruits_list[fruit_id]
        
    raise RuntimeError(f"Fruit with id {fruit_id} does not exist")
```

```python
import numpy as np

def swap(coords: np.ndarray) -> np.ndarray:
    # Safely create a new array with columns rearranged to: [col 1, col 0, col 3, col 2, col 4]
    swapped = coords[:, [1, 0, 3, 2, 4]]
    return swapped
```

```python
!pip install torch torchvision

import os
from typing import Tuple
from PIL import Image
import torch
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, img_dir, transform=None):
        self.img_dir = img_dir
        self.img_names = sorted(os.listdir(img_dir))
        self.img_paths = [os.path.join(img_dir, name) for name in self.img_names]
        self.transform = transform

    def __len__(self) -> int:
        return len(self.img_paths)

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, str]:
        img_path = self.img_paths[idx]

        with Image.open(img_path) as img:
            img = img.convert('RGB')

            if self.transform is not None:
                img = self.transform(img)

        return img, img_path
```
